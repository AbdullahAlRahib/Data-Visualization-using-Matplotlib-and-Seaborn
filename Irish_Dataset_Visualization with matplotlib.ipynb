Histogram

```python
# View the histogram of "SepalLengthCm" column
plt.hist(x=data['SepalLengthCm'])
plt.title('Histogram of "SepalLengthCm" column')
plt.xlabel('SepalLength (cm)')
plt.ylabel('Quantity')
plt.show()
```

```python
#View all the culumns using the same histgram plotting
plt.hist(x=[
    data.iloc[:,1],
    data.iloc[:,2],
    data.iloc[:,3],
    data.iloc[:,4],
])
plt.show()
```

```python
#View all the culumns using the same histgram plotting
#Display the column names as legend in the plot
plt.hist(
    x=[data.iloc[:,1],data.iloc[:,2],data.iloc[:,3],data.iloc[:,4],],
    label=(data.columns[1],data.columns[2],data.columns[3],data.columns[4])
)
plt.legend() 
plt.show()
```

```python
# View histogram plotting of all numeric columns
# Overlap on each other , set diffrent opacity levels
plt.hist(data['SepalLengthCm'])
plt.hist(data['SepalWidthCm'], alpha=0.7)
plt.hist(data['PetalLengthCm'],alpha=0.5)
plt.hist(data['Species'],alpha=0.2)
plt.show()
```

```python
#View all the hist Plotting seperately in different subplots
#Define the figure-sizes of each subplots
#Since there will $ subplots, thus they'll divided into 2 rows & cols
fig, ax =plt.subplots(nrows=2, ncols=2, figsize=(10,8))

ax[0,0].hist(data['SepalLengthCm'])
ax[0,0].set_title('SepalLengthCm')
ax[0,0].set_xlabel('Cm')
ax[0,0].set_ylabel('Count')

ax[0,1].hist(data['SepalWidthCm'])
ax[0,1].set_title('SepalWidthCm')
ax[0,1].set_xlabel('Cm')
ax[0,1].set_ylabel('Count')

ax[1,0].hist(data['PetalLengthCm'])
ax[1,0].set_title('PetalLengthCm')
ax[1,0].set_xlabel('Cm')
ax[1,0].set_ylabel('Count')

ax[1,1].hist(data['PetalWidthCm'])
ax[1,1].set_title('PetalWidthCm')
ax[1,1].set_xlabel('Cm')
ax[1,1].set_ylabel('Count')

plt.subplots_adjust(top = 0.99, bottom=0.01, hspace=.5 ,wspace=0.4)
plt.show()
```

**Scatter plot**

```python
# Make a scatter plot on three species on their "SepalWidthCm" & "petalLengthCm"
# The "Species" column will be passed througt the "color" param("c")
color = {
    'Iris-setosa':'black',
    'Iris-versicolor':'blue',
    'Iris-virginica':'green',
}
plt.scatter(
    x=data['PetalLengthCm'],
    y=data['SepalLengthCm'],
    c=data['Species'].map(color)
)
plt.xlabel('PetalLengthCm')
plt.ylabel('SepalLengthCm')
plt.title('Iris Data Scatter Plot')
plt.show()
```

```python
#Got all the dataframes on species
df_setosa = data[data['Species'] == 'Iris-setosa']
df_setosa_values = df_setosa.loc[:,'PetalWidthCm'].values
df_versicolor = data[data['Species'] == 'Iris-versicolor']
df_versicolor_values = df_versicolor.loc[:,'PetalWidthCm'].values
df_virginica = data[data['Species'] == 'Iris-virginica']
df_virginica_values = df_virginica.loc[:,'PetalWidthCm'].values
```

```python
# Petal Width
df = pd.DataFrame({
    'df_setosa' : list(df_setosa_values),
    'df_versicolor' : list(df_versicolor_values),
    'df_virginica' : list(df_virginica_values),
})
ax =df.plot.area()
plt.xlabel('Quantity')
plt.ylabel('Petal Width (Cm)')
plt.show()
```

```python
#Sepal Length (pie Plot)
labels = data_grouped.index.to_list()
sepal_size = list(data_grouped.SepalLengthCm.values)
plt.pie(sepal_size, labels=labels,  autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')
fig=plt.gcf()
fig.set_size_inches(10,10)
plt.title("Sepal Length (AVG cm)")
plt.show()
```
