# Histogram

A histogram is an accurate representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable. It is a kind of bar graph.
1.   **Its allows only numerical data**




**Matplotlib**

```python
# View the histogram of "Pclass" column
plt.hist(x=data['Survived'])
plt.title('Histogram of "Survived" column')
plt.xlabel('Survived')
plt.ylabel('Quantity')
plt.show()
```

```python
# View the histogram of "Sex" column
plt.hist(x=data['Sex'])
plt.title('Histogram of "Sex" column')
plt.xlabel('Sex')
plt.ylabel('Quantity')
plt.show()
```

```python
#View all the culumns using the same histgram plotting
plt.hist(x=[
    data.iloc[:,1],
    data.iloc[:,2],
    data.iloc[:,6],
    data.iloc[:,7],
])
plt.show()
```

```python
#View all the culumns using the same histgram plotting
#Display the column names as legend in the plot
plt.hist(x=[data.iloc[:,1],data.iloc[:,2],data.iloc[:,6],data.iloc[:,7]],  
    label=(data.columns[1],data.columns[2],data.columns[6],data.columns[7]))
plt.legend() # To Show those Labels
plt.show()
```

```python
# View histogram plotting of all numeric columns
# Overlap on each other , set diffrent opacity levels
plt.hist(data['Survived'])
plt.hist(data['Pclass'], alpha=0.7)
plt.hist(data['SibSp'],alpha=0.5)
plt.hist(data['Parch'],alpha=0.2)
plt.show()
```

```python
#View all the hist Plotting seperately in different subplots
#Define the figure-sizes of each subplots
#Since there will $ subplots, thus they'll divided into 2 rows & cols
fig, ax =plt.subplots(nrows=2, ncols=2, figsize=(10,8))

ax[0,0].hist(data['Survived'])
ax[0,0].set_title('Survived')
ax[0,0].set_xlabel('type')
ax[0,0].set_ylabel('Count')

ax[0,1].hist(data['Pclass'])
ax[0,1].set_title('Pclass')
ax[0,1].set_xlabel('Class type')
ax[0,1].set_ylabel('Count')

ax[1,0].hist(data['SibSp'])
ax[1,0].set_title('SibSp')
ax[1,0].set_xlabel('Sibsp type')
ax[1,0].set_ylabel('Count')

ax[1,1].hist(data['Parch'])
ax[1,1].set_title('Parch')
ax[1,1].set_xlabel('Parch varaites')
ax[1,1].set_ylabel('Count')

plt.subplots_adjust(top = 0.99, bottom=0.01, hspace=.5 ,wspace=0.4)
plt.show()
```

# Scatter plot



You can use scatter plots to explore the relationship between two variables, for example by looking for any correlation between them. 

```python
data.head()
```

```python
# Make a scatter plot on three species on their "SepalWidthCm" & "petalLengthCm"
# The "Species" column will be passed througt the "color" param("c")
color = {
    'male':'black',
    'female':'blue',
}
plt.scatter(
    x=data['Age'],
    y=data['Fare'],
    c=data['Sex'].map(color)
)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Titanic Scatter Plot')
plt.show()
```

# Box Plot

It divides the data set into three quartiles. This graph represents the minimum, maximum, median, first quartile and third quartile in the data set. It is also useful in comparing the distribution of data across data sets by drawing boxplots for each of them.
1.   Also helps to find Outliers








```python
plt.boxplot([
    data['Fare'],
])
plt.show()
```

```python
#View the spread of data though different sexttype
data.boxplot(column="Fare", by='Sex')
```

```python
# View the average age based on 'Pclass' columns
sns.boxplot(x='Pclass', y='Age', data=data, palette='hls')
plt.show()
```

# Bar Plot

Histograms and bar charts can both display large sets of data But
*   **Bar charts display categorical data**
*   **information that's separated into different groups based on
characteristics**, while histograms display numerical or quantitative data in 
bins, which is data that you can measure with numbers.

```python
d=data.Pclass.value_counts()
d=d.to_dict()
i=d.keys()
v=d.values()
plt.bar(i, v)
plt.title("Pclass Values count")
plt.show()
```

# Area Plot

```python
df = pd.DataFrame({
    'sales' : [3,2,3,9,10,6],
    'signups' : [5,5,6,12,14,13],
    'visits' : [20, 42, 28,62,81,50],
}, index= pd.date_range(start='2018/01/01', end='2018/07/01',
                            freq='M'))
ax= df.plot.area()
```

# Pie Chart

```python
labels=['male','female']
sex=list(data.groupby(['Sex'])['Sex'].count())#Here we grouped by sex by sex
plt.pie(sex, labels=labels,autopct='%1.1f%%',shadow=True, startangle=90)
plt.axis('equal')
fig=plt.gcf()
fig.set_size_inches(10,10)
plt.title("Petal Width(AVG cm)")
plt.show()
```
